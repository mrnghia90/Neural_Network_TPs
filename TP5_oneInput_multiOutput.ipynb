```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Model
from keras.layers import Dense
from keras import Input
import tensorflow as tf
```

```python
#fix random seed-1 (or whatever other integer) for reproductibility
seed = 7
np.random.seed(seed)
```

```python
#Generate data
N = 1000
X_train = np.linspace(0.0,1.0,N)
y_train1 = np.sin(2*np.pi*X_train)
y_train2 = np.exp(X_train)

X_val = np.random.rand(100)
X_val = np.sort(X_val)
y_val1 = np.sin(2*np.pi*X_val)
y_val2 = np.exp(X_val)
```

```python
#Define model
x0 = Input(shape = (1,))

x1 = Dense(20,activation = 'sigmoid',kernel_initializer = "uniform")(x0)
x2 = Dense(20, activation = 'sigmoid', kernel_initializer = "uniform")(x1)

x3_left = Dense(20,activation = 'sigmoid',kernel_initializer = "uniform")(x2)
x3_right = Dense(30, activation = 'sigmoid', kernel_initializer = "uniform")(x2)

x4_left = Dense(20,activation = 'sigmoid',kernel_initializer = "uniform")(x3_left)
x4_right = Dense(30, activation = 'sigmoid', kernel_initializer = "uniform")(x3_right)

x5_left = Dense(1,kernel_initializer = "uniform")(x4_left)
x5_right = Dense(1,kernel_initializer = "uniform")(x4_right)
                                                                             
model = Model(inputs = x0, outputs = [x5_left,x5_right])
model.compile(optimizer = 'rmsprop', loss = ['mse','mse'],metrics = 'mse')
hist = model.fit(X_train,[y_train1,y_train2],epochs = 300, batch_size = 32,validation_data = (X_val,[y_val1,y_val2]))
    
```

```python
train_loss = hist.history['loss']
valid_loss = hist.history['val_loss']
ep = range(1, len(train_loss)+1)
plt.figure()
plt.plot(ep,train_loss,'b-')
plt.plot(ep,valid_loss,'r-')
plt.title('Training and validation loss')
plt.show()
```

```python
tf.keras.utils.plot_model(model,show_shapes = True)
```

```python
[y_sin,y_exp] = model.predict(X_val)
```

```python
plt.figure(figsize = (7,7))
plt.plot(X_val,y_sin,'b')
plt.plot(X_val,y_val1,'r')
plt.title('Trainning and validation data for output 1')
plt.show()
```

```python
plt.figure(figsize = (7,7))
plt.plot(X_val,y_exp,'b')
plt.plot(X_val,y_val2,'r')
plt.title('Trainning and validation data for output 2')
plt.show()
```
