```python
%reset -f
import numpy as np
import time
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras.layers import Dense

import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

```python
np.random.seed(1)
```

```python
#Read the data
dataset = np.loadtxt('pima-indians-diabetes.csv', delimiter = ",")
print(dataset.shape)
print(dataset)
```

```python
#Split int to input(X) and output (Y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
def creat_model(X,y):

    model = Sequential()

    # No choice for the input dimension, it must be 8 because of 8 parameters
    model.add(Dense(12,activation = 'relu',input_dim = 8))    
    model.add(Dense(8, activation = 'relu', kernel_initializer = "uniform"))
    model.add(Dense(1, kernel_initializer = "uniform", activation = 'sigmoid'))
    
    model.compile(optimizer = 'adam', loss = 'binary_crossentropy',metrics =  'accuracy')
    model.fit(X,y,epochs = 100, batch_size = 10,validation_split = 0.33)
    
    return model
```

```python
from sklearn.model_selection import StratifiedKFold
# define 10-fold data validation
kfold = StratifiedKFold(n_splits = 10, shuffle = True, random_state = 0)

cv_score = []

for train_index,test_index in kfold.split(X,y):
    model = creat_model(X[train_index],y[train_index])
    
    score = model.evaluate(X[test_index],y[test_index],verbose  = 0)
    print(model.metrics_name[1],score[1]*100)
    cv_score.append(score[1]*100)
    
print('average accuracy and standard deviation:', np.mean(cv_score), np.std(cv_score))
```
